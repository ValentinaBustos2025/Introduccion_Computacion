```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Funciones
#Punto 1
def f_(n):
    return n*math.log2(n)

def n_cuadrado(n):
    return n**2

def dos_n(n):
    return 2**n

#Punto 2
def f(n):
    return 2 ** np.sqrt(n) + n ** 12

def g(n):
    return 3 ** (n ** (1/3)) + n ** 17

def h(n):
    log_term = np.log(n**7 + 2) / np.log(5)
    return log_term ** 19 + n ** 23
```

```python
#DatSet 
n_vals = np.linspace(2, 50, 100)  

#Punto 1
f_vals_prim = [f(n) for n in n_vals]
dos_n_vals = [dos_n(n) for n in n_vals]
n_cuadrado_vals = [n_cuadrado(n) for n in n_vals]


#Punto 2 
f_vals = [f(n) for n in n_vals]
g_vals = [g(n) for n in n_vals]
h_vals = [h(n) for n in n_vals]
```

```python
plt.style.use('default')
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(n_vals, f_vals_prim, label='f(n) = n log2 n', linewidth=2)
ax.plot(n_vals, dos_n_vals, label='n^2', linewidth=2)
ax.plot(n_vals, n_cuadrado_vals, label='2^(n)', linewidth=2)

ax.set_xlabel('n')
ax.set_ylabel('Valor (escala logarítmica)')
ax.set_title('Comparación de crecimiento de f(n) en comparación')
ax.set_yscale('log')  
ax.legend()
ax.grid(True, which="both", linestyle='--', linewidth=0.5)

axins = ax.inset_axes([0.55, 0.45, 0.35, 0.35])  

# Filtrar datos para x entre 30 y 40
n_vals_array = np.array(n_vals)
f_vals_prim_array = np.array(f_vals_prim)
dos_n_vals_array = np.array(dos_n_vals)
n_cuadrado_vals_array = np.array(n_cuadrado_vals)

mask = (n_vals_array >= 20) & (n_vals_array <= 35)
n_vals_detalle = n_vals_array[mask]
f_vals_prim_detalle = f_vals_prim_array[mask]
dos_n_vals_detalle = dos_n_vals_array[mask]
n_cuadrado_vals_detalle = n_cuadrado_vals_array[mask]

# Graficar en el inset (escala real)
axins.plot(n_vals_detalle, f_vals_prim_detalle, linewidth=2)
axins.plot(n_vals_detalle, dos_n_vals_detalle, linewidth=2)
axins.plot(n_vals_detalle, n_cuadrado_vals_detalle, linewidth=2)


axins.set_ylim(auto=True)  
axins.set_xlabel('n', fontsize=8)
axins.set_ylabel('Valor real', fontsize=8)
axins.tick_params(axis='both', labelsize=7)
axins.grid(True, linestyle='--', linewidth=0.5)


plt.tight_layout()
plt.show()
```
